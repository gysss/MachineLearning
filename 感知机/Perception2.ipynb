```python
from sklearn.preprocessing import LabelEncoder

class_label = LabelEncoder()
col = ['b', 'd', 'f', 'g', 'h', 'i', 'j', 'n']
for i in col:
    df[i] = class_label.fit_transform(df[i].values)
```

```python
df.label[df['label'] == ' <=50K'] = -1
df.label[df['label'] == ' >50K'] = 1
df.head()
```

```python
df.info()
```

```python
import numpy as np
col = ['a', 'b', 'd', 'e','f','g','h','i','j','m','n','label']
pd = df.loc[:, col]
dataset = np.array(pd,dtype = np.float32)
print(dataset[0])
x_train, y_train = dataset[:25000, :-1], dataset[:25000,-1]
x_test, y_test = dataset[25000:, :-1], dataset[25000:,-1]
print(x_train[0])
print(y_train[0])
```

```python
class Model:
    def __init__(self):
        self.w = np.ones(len(dataset[0] )- 1, dtype=np.float32)
        self.b = 0
        self.l_rate = 0.5
    
    def sign(self, x, w, b):
        y = np.dot(x, w) + b
        return y
    
    # 随机梯度下降
    def fit(self, xtrain, ytrain, max_iter = 100):
        while max_iter>=0:
            flag = 0
            for i in range(len(xtrain)):
                x = xtrain[i]
                y = ytrain[i]
                if y * self.sign(x, self.w, self.b) <= 0:
                    self.w = self.w + self.l_rate * np.dot(y, x)
                    self.b = self.b + self.l_rate * y
                    flag = 1
            if flag == 0:
                break
            max_iter -= 1
        return "Perceptron Finished!"
    
    def score(self, x_test, y_test):
        right_count = 0
        for i in range(len(x_test)):
            x = x_test[i]
            y = y_test[i]
            if y * self.sign(x, self.w, self.b) > 0:
                right_count+=1
        return right_count / len(x_test)
```

```python
perceptron = Model()
perceptron.fit(x_train, y_train, max_iter= 100)
```

```python
# 测试集上的准确率
perceptron.score(x_test, y_test)
```

```python
# 训练集上的准确率
perceptron.score(x_train, y_train)
```

### 交叉验证

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import cross_val_score, ShuffleSplit
```

```python
import numpy as np
col1 = ['a', 'b','c','d', 'e','f','g','h','i','j','k','l','m','n','label']
pd = df.loc[:, col1]
dataset3 = np.array(pd,dtype = np.float32)
```

```python
clf = Perceptron(fit_intercept=False, max_iter=100, shuffle=False)
x_train_cv, y_train_cv=dataset3[:, :-1], dataset3[:,-1] 
cv = ShuffleSplit(n_splits=10, test_size = 0.18, random_state=0)
accs = cross_val_score(clf, x_train_cv, y_train_cv, scoring = None, cv = cv)
print(accs)
```
